### Ý nghĩa của các dòng trong dữ liệu

Mỗi dòng trong dữ liệu cung cấp các thông tin liên quan đến bệnh nhân. Gồm các yếu tố như:

| **Cột**   | **Ý nghĩa (Tiếng Anh)**| **Ý nghĩa (Tiếng Việt)**|
|-----------|------------------------|-------------------------|
| **id**    | | Mã định danh duy nhất   |
| **gender**| "Male", "Female" or "Other"| "Nam", "Nữ" hoặc "Khác"|
| **age**   |                            | Tuổi của bệnh nhân     |
| **hypertension** | | 0 nếu bệnh nhân không bị tăng huyết áp, 1 nếu bệnh nhân bị tăng huyết áp   |
| **heart_disease** | | 0 nếu bệnh nhân không mắc bệnh tim, 1 nếu bệnh nhân mắc bệnh tim |
| **ever_married**    | "No" or "Yes" | "Không" hoặc "Có" |
| **work_type**       | "children", "Govt_job", "Never_worked", "Private" or "Self-employed"| "Trẻ em", "Việc làm nhà nước", "Chưa từng làm việc", "Tư nhân" hoặc "Tự kinh doanh"    |
| **Residence_type**  | "Rural" or "Urban" | "Nông thôn" hoặc "Thành thị"                                                           |
| **avg_glucose_level** | | Mức đường huyết trung bình                                                             |
| **bmi**             | Body Mass Index  | Chỉ số khối cơ thể                                                                      |
| **smoking_status**  | "formerly smoked", "never smoked", "smokes" or "Unknown" | "Đã từng hút thuốc", "Chưa từng hút thuốc", "Đang hút thuốc" hoặc "Không xác định"*    |
| **stroke**          | | 1 nếu bệnh nhân bị đột quỵ, 0 nếu không |

#### Ghi chú:
- Các giá trị "Unknown" trong smoking_status nghĩa là không có thông tin cho bệnh nhân đó

### Dữ liệu có bị trùng lắp không
